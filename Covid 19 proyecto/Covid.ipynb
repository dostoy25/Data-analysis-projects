## 3 Exploracion de datos
Comenzaremos a explorar los datos ya que se limpiaron lo suficiente como para manejarlos a nuestra manera.

### 3.1 ¿Cuantos casos y muertos hubieron en total  registrados?

```python
%%sql
SELECT MAX(total_cases) AS maximo_de_casos, MAX(total_deaths) maximo_de_muertos
FROM covid
```

### 3.2 ¿Cual fue el pais con mayor numero de casos?

```python
%%sql
SELECT country, MAX(total_cases) AS mayor_numero_casos
FROM covid
WHERE continent IS NOT NULL
GROUP BY country
ORDER BY mayor_numero_casos DESC
LIMIT 50
```

### 3.3 ¿Cual fue el pais con mayor numero de muertes registradas?

```python
%%sql
SELECT country, MAX(total_deaths) AS muertos_totales
FROM covid
WHERE continent IS NOT NULL
GROUP BY country
ORDER BY muertos_totales DESC
LIMIT 50
```

### 3.4 ¿Cual fue el porcentaje de muertos en base a sus casos?

```python
%%sql
SELECT country, date_covid, total_cases, total_deaths, (total_deaths/total_cases)*100 AS muertos_casos_porcentaje
FROM covid
ORDER BY 1
LIMIT 50
```

### 3.5 ¿Cual fue el porcentaje de infectados en base a su poblacion?

```python
%%sql
SELECT DISTINCT country, date_covid, total_cases, population, (total_cases/population)*100 AS casos_porcentaje
FROM covid
WHERE continent IS NOT NULL
LIMIT 50
```

### 3.6 ¿Cual fue el porcentaje de muertos en base a su poblacion?

```python
%%sql
SELECT DISTINCT country, date_covid, total_cases, total_deaths, (total_deaths/total_cases)*100 AS muertos_casos_porcentaje
FROM covid
LIMIT 50
```

### 3.7 ¿Cuales fueron  los mayores casos en Estados Unidos?

```python
%%sql
SELECT country, date_covid, MAX(new_cases) AS mayor_numero_casos 
FROM covid
WHERE country LIKE '%States%' 
GROUP BY country, date_covid, total_cases
ORDER BY mayor_numero_casos
LIMIT 50
```

### 3.8 ¿Cual fue el porcentaje de muertos en estados unidos en base a sus casos?

```python
%%sql
SELECT DISTINCT country, date_covid, total_cases, total_deaths, (total_deaths/total_cases)*100 AS muertos_casos_porcentaje
FROM covid
WHERE country ILIKE '%states%'
LIMIT 50
```

### 3.9 ¿Cual fue el porcentaje de personas muertas en base a su poblacion?

```python
%%sql
SELECT DISTINCT country, date_covid, total_deaths, (total_deaths/population)*100 AS porcentaje_muertos
FROM covid
LIMIT 50
```

### 3.10 ¿Cuales fueron los continentes con el numero total mas alto de muertos?

```python
%%sql
SELECT continent,  SUM(new_deaths) AS total_muertos
FROM covid
WHERE continent IS NOT NULL
GROUP BY continent
ORDER BY total_muertos DESC
LIMIT 3
```

### 3.11 ¿Cuales fueron los continente con el numero total mas alto de casos?

```python
%%sql
SELECT continent,  SUM(new_cases) AS total_casos
FROM covid
WHERE continent IS NOT NULL
GROUP BY continent
ORDER BY total_casos DESC
LIMIT 3
```

### 3.12 ¿Se puede encontrar la manera de tener la vacunacion en la fecha anterior y posterior?

```python
%%sql
WITH paises_vacunados AS (SELECT country, date_covid, MAX(total_vaccinations) AS maximo
						  FROM covid
						  WHERE continent IS NOT NULL
						  GROUP BY country, date_covid
						  HAVING MAX(total_vaccinations) IS NOT NULL),
						  pais_acumulativos AS (SELECT country, date_covid, maximo,
						  						LAG(maximo, 1) OVER(ORDER BY maximo DESC) AS posterior
						  						FROM paises_vacunados)
SELECT country, date_covid, maximo, posterior
FROM pais_acumulativos
ORDER BY 1, 2
LIMIT 150
```

### 3.13 ¿Cual es el continente con la mayor cantidad de personas vacunadas?

```python
%%sql
SELECT DISTINCT continent, MAX(people_vaccinated) AS personas_vacunadas
FROM covid
WHERE people_vaccinated IS NOT NULL AND continent IS NOT NULL
GROUP BY continent
ORDER BY personas_vacunadas DESC
LIMIT 1
```

### 3.14 ¿Se puede relacionar el GDP per capita?

```python
%%sql
SELECT DISTINCT country, MAX(people_vaccinated_per_hundred_new) AS personas_vacunadas_cada_100, MAX(gdp_per_capita_new) AS maximo_gdp_per_capita
FROM covid
WHERE gdp_per_capita_new IS NOT NULL AND people_vaccinated_per_hundred_new IS NOT NULL
GROUP BY country
HAVING MAX(people_vaccinated_per_hundred_new) < 100
ORDER BY personas_vacunadas_cada_100 DESC, maximo_gdp_per_capita DESC
LIMIT 50
```

### 3.15 ¿Cual es la diferencia entre personas totalmente vacunadas y las que solo estan vacunadas?

```python
%%sql
SELECT country, MAX(people_vaccinated) AS personas_vacunadas, MAX(people_fully_vaccinated) AS personas_totalmente_vacunadas
FROM covid
GROUP BY country
ORDER BY 1
LIMIT 50
```

### 3.16 ¿Cual es el pais con el mayor numero de personas totalmente vacunados?

```python
%%sql
SELECT country, MAX(people_fully_vaccinated) AS totalmente_vacunados
FROM covid
WHERE people_fully_vaccinated IS NOT NULL AND continent IS NOT NULL
GROUP BY country
ORDER BY totalmente_vacunados DESC
LIMIT 1
```

### 3.17 ¿Cuales paises superan el porcentaje de ingresados a cuidados intensivos?

```python
%%sql
SELECT DISTINCT country, date_covid, icu_patients
FROM covid
WHERE icu_patients > (SELECT AVG(icu_patients)
					 FROM covid)
LIMIT 20
```

### 3.18 ¿Hay una relacion en cuanto la admision semanal de pacientes y los pacientes que ya estaban?

```python
%%sql
SELECT DISTINCT country, MAX(weekly_hosp_admissions) AS maximo_pacientes_ingresados, MAX(hosp_patients) AS maximo_pacientes
FROM covid
GROUP BY country
HAVING MAX(hosp_patients) IS NOT NULL AND MAX(weekly_hosp_admissions) IS NOT NULL
ORDER BY maximo_pacientes_ingresados, maximo_pacientes
```

### 3.19 Creacion de vista para visualizadores como: power bi, tableau, etc.

```python
%%sql
CREATE VIEW fecha_anterior_posterior AS
WITH paises_vacunados AS (SELECT country, date_covid, MAX(total_vaccinations) AS maximo
						  FROM covid
						  WHERE continent IS NOT NULL
						  GROUP BY country, date_covid
						  HAVING MAX(total_vaccinations) IS NOT NULL),
						  pais_acumulativos AS (SELECT country, date_covid, maximo,
						  						LAG(maximo, 1) OVER(ORDER BY maximo DESC) AS posterior
						  						FROM paises_vacunados)
SELECT country, date_covid, maximo, posterior
FROM pais_acumulativos
ORDER BY 1, 2
LIMIT 150
```
